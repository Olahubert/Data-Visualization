```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load dataset
df = pd.read_csv('Netflix+Dataset.csv')
```

```python
df
```

```python
#Check for duplicate records and remove them
duplicate_records = df.duplicated().sum()
duplicate_records
```

```python
#Check for null values
null_values = df.isnull().sum()
```

```python
null_values
```

```python
#Show Id and Director of 'House of Cards'
house_of_cards = df[df['Title'] == 'House of Cards'][['Show_Id', 'Director']]
```

```python
house_of_cards
```

```python
#Find the year with the highest number of TV Shows & Movies released
year_counts = df['Release_Date'].value_counts()
year_with_highest_releases = year_counts.idxmax()
```

```python
year_counts
```

```python
year_with_highest_releases
```

```python
#Plotting the bar graph for the number of releases per year
year_counts.plot(kind='bar', figsize=(12, 6))
plt.title('Number of TV Shows & Movies Released per Year')
plt.xlabel('Year')
plt.ylabel('Number of Releases')
plt.show()
```

```python
#Count the number of Movies and TV Shows and show with bar graph
category_counts = df['Category'].value_counts()
```

```python
category_counts
```

```python
category_counts.plot(kind='bar', figsize=(6, 6))
plt.title('Number of Movies & TV Shows')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
#Show all the Movies that were released in the year 2000
movies_2000 = df[(df['Release_Date'] == 2000) & (df['Category'] .str.contains('Movie'))]
```

```python
movies_2000
```

```python
#Show only the titles of TV Shows that were released in India only
tv_shows_india = df[(df['Category'] == 'TV Show') & (df['Country'] == 'India')]['Title']
```

```python
tv_shows_india
```

```python
 #Top 10 Directors with the highest number of TV Shows & Movies on Netflix
top_10_directors = df['Director'].value_counts().head(10)
top_10_directors
```

```python
#Show all records where Category is Movie and Type is Comedies or Country is United Kingdom
movies_comedies_uk = df[((df['Category'] == 'Movie') & (df['Type'].str.contains('Comedies'))) | (df['Country'] == 'United Kingdom')]
```

```python
movies_comedies_uk
```

```python
#Count the number of movies/shows Tom Cruise was cast in
tom_cruise_count = df[df['Cast'].str.contains('Tom Cruise', na=False)].shape[0]
tom_cruise_count
```

```python
#Different ratings defined by Netflix
different_ratings = df['Rating'].unique()
different_ratings
```

```python
#Number of Movies that got the 'TV-14' rating in Canada
tv14_canada_movies = df[(df['Rating'] == 'TV-14') & (df['Country'] == 'Canada') & (df['Category'] == 'Movie')].shape[0]
tv14_canada_movies
```

```python
#Number of TV Shows that got the 'R' rating after the year 2018
r_rating_tv_shows_post_2018 = df[(df['Rating'] == 'R') & (df['Type'] == 'TV Show') & (df['Release_Date'] > 2018)].shape[0]
r_rating_tv_shows_post_2018
```

```python
#Maximum duration of a Movie/Show on Netflix
max_duration = df['Duration'].max()
max_duration
```

```python
#Country with the highest number of TV Shows
country_tv_shows = df[df['Category'] == 'TV Show']['Country'].value_counts().idxmax()
country_tv_shows 
```

```python
# Sort the dataset by Year
sorted_df = df.sort_values(by='Release_Date')
sorted_df
```
