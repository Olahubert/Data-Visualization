```python
import matplotlib.pyplot as plt
```

```python
pd.read_csv("sales.csv")
```

```python
Sales=pd.read_csv("sales.csv")
```

```python
Sales.head(20)
```

```python
Sales1=Sales.head(20)
```

```python
Sales.tail(20)
```

```python
Sales2=Sales.tail(20)
```

```python
Sales1["Total Expenses"]
```

```python
Total_expenses1=Sales1["Total Expenses"]
```

```python
Sales1["Product Type"]
```

```python
Product_type1=Sales1["Product Type"]
```

```python
Sales2["Total Expenses"]
```

```python
Total_expenses2=Sales2["Total Expenses"]
```

```python
Sales2["Product Type"]
```

```python
Product_type2=Sales2["Product Type"]
```

```python
fig, ax = plt.subplots(2,figsize =(20,20))     # Create a figure and a grid of subplots with 2 rows
fig.suptitle('Sale/Revenue', fontsize=14, fontweight='bold')  
ax[0].pie(Total_expenses1, labels=Product_type1, autopct="%.2f%%")
ax[1].pie(Total_expenses2, labels=Product_type2, autopct="%.2f%%")

ax[0].set_title("Total_expenses/Product_type1")
ax[1].set_title("Total_expenses2/Product_type2")

plt.show()                                                
```
